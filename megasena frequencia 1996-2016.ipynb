```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
megasena = pd.read_csv('megasena.csv')
```

```python
#criando um df com apenas os resultados e numero do concurso
colunas_relevantes = ['1ª Dezena','2ª Dezena','3ª Dezena','4ª Dezena','5ª Dezena','6ª Dezena']
megasena_colunas = megasena[colunas_relevantes]
#aqui retornamos os valores e formatamos a info do dataframe em array unidimensional pra que consiga fazer o grafico
colunas_formatadas = megasena_colunas.values.flatten()
```

```python
#obtendo a frequencia de cada numero sorteado
frequencia = pd.Series(colunas_formatadas).value_counts().sort_index()
frequencia
```

```python
#criando um dataframe de frequencia
df_freq = pd.DataFrame({'Dezena': frequencia.index, 'Frequência': frequencia.values})
#plotando o histograma de frequencia
plt.figure(figsize=(24, 12))
sns.barplot(x='Dezena', y='Frequência', data=df_freq, hue='Dezena', palette='viridis', dodge=False, legend=False)
plt.xlabel('Dezenas')
plt.ylabel('Frequência')
plt.title('Frequência de cada dezena sorteada na Mega-Sena')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

```python
#selecionando os 6 mais frequentes
top_6 = df_freq.nlargest(6,'Frequência')

#grafico
plt.figure(figsize=(6, 3))
sns.barplot(x='Dezena', y='Frequência', data=top_6, hue='Dezena', palette='viridis', dodge=False, legend=False)
plt.xlabel('Dezenas')
plt.ylabel('Frequência')
plt.title('Frequência de cada dezena sorteada na Mega-Sena')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```
